# Popularity Based Recommender System

```python
print(books.shape)
books = books.dropna(axis=0, how='any')
print(books.shape)
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings.shape
```

```python
ratings_with_books.groupby('Book-Title').count()['Book-Rating']
```

```python
ratings_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df = ratings_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'count_of_ratings'},inplace=True)
```

```python
num_rating_df.head(7)
```

```python
avg_rating_df = ratings_with_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
# extracting top 50 books based on the average rating
# where number of ratings must be greater than 250
popular_df[popular_df['count_of_ratings'] >= 250]
```

```python
popular_df = popular_df[popular_df['count_of_ratings'] >= 250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df.head()
```

```python
popular_df.merge(books,on='Book-Title').head()
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df = popular_df[['Book-Title','Book-Author','Image-URL-M','count_of_ratings','avg_ratings']]
```

```python
popular_df.head()
```

```python
# df.round(2) --> will round all the numericals column upto three place
# # round off the columns in this manner
# "A" to 1 decimal place
# "B" to 2 decimal place
# "C" to 3 decimal place
# "D" to 4 decimal place
# df.round({"A":1, "B":2, "C":3, "D":4})

popular_df = popular_df.round({'avg_ratings':2})
```

```python
popular_df = popular_df.reset_index(drop=True)
popular_df
```

```python
popular_df
```

```python
# store_df = popular_df

popular_df['Image-URL-M'] = popular_df['Image-URL-M'].apply(lambda x: x[0:4] + 's' + x[4:-1]) 
```

```python
popular_df
```

```python
popular_df.to_csv('popular_books.csv',index=False)
```
