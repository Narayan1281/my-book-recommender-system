# # Collaborative Filtering Based Recomender System


```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv('Books.csv')
users = pd.read_csv('Users.csv')
ratings = pd.read_csv('Ratings.csv')
```

```python
# books.isnull().sum()
books.dropna(axis=0, how='any',inplace=True)
```

```python
books['Image-URL-M'].describe()
```

```python
ratings_join_books = ratings.merge(books, on='ISBN')
```

```python
ratings.shape
```

```python
ratings_join_books.head()
```

```python
var = ratings_join_books.groupby('User-ID').count()['Book-Rating'] >=200
```

```python
var.describe()
```

```python
experienced_users = var[var].index
```

```python
filtered_ratings_join_books = ratings_join_books[ratings_join_books['User-ID'].isin(experienced_users)]
filtered_df = filtered_ratings_join_books
```

```python
# similarly filtering books based on number of ratings
y = filtered_df.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
print(filtered_df.shape)
filtered_df.drop_duplicates()
print(filtered_df.shape)
```

```python
final_df = filtered_df[filtered_df['Book-Title'].isin(famous_books)]
```

```python
final_df
```

```python
final_df['Image-URL-M'] = final_df['Image-URL-M'].apply(lambda x: x[0:4] + 's' + x[4:-1]) 
```

```python
final_df.to_csv('dump_books.csv')
```

```python
new_dump = pd.read_csv('dump_books.csv')
new_dump
```

```python
pt = final_df.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt
```

```python
pt.describe()
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# We shall use cosine_similarity as a measure to find similarity as
# pivot table has a large number of NaN values
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
# make a DataFrame out of similarity_score --> will be easier to save as a csv file
similarity_score = pd.DataFrame(similarity_score)
```

```python
similarity_score
```

```python
similarity_score.to_csv('dump_similarity_scores.csv',index=False)
```

```python
check = pd.read_csv('dump_similarity_scores.csv')
check
```

```python
def recommend(book_name):
    # fetch index
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(list(similarity_score.iloc[index]))), key=lambda x:x[1], reverse=True)[1:8]
    # we shall show 7 recommendation based on the given book_name
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title']==pt.index[i[0]]].drop_duplicates('Book-Title')
        item.append(temp_df['Book-Title'].values[0])
        item.append(temp_df['Book-Author'].values[0])
        item.append(temp_df['Image-URL-M'].values[0])
        
        data.append(item)
    return data
        
    
```

```python
recommend('Message in a Bottle')
```
